**Node Class**
 This class represents a node in the Huffman tree.
    Each node has a frequency (`freq`), a symbol (`symbol`),
    and optionally left and right child nodes (`left` and `right`).
    The `__lt__` method is used to compare nodes based on their frequency.


```python
class Node:
    def __init__(self, freq, symbol, left=None, right=None):
        self.freq = freq
        self.symbol = symbol
        self.left = left
        self.right = right
        self.huff = ''

    # Definiowanie operatora < dla kolejki priorytetowej
    def __lt__(self, other):
        return self.freq < other.freq
```

**PriorityQueue Class**
This class is a simple implementation of a priority queue using a list and the heapq module. It provides methods to insert a node (insert), extract the node with the minimum frequency (extract_min), and check if the queue is empty (is_empty).

```python
class PriorityQueue:
    def __init__(self):
        self.queue = []

    def insert(self, node):
        heapq.heappush(self.queue, node)

    def extract_min(self):
        return heapq.heappop(self.queue)

    def is_empty(self):
        return len(self.queue) == 0
```

**create_huffman_tree** This function takes a sequence of symbols and creates a Huffman tree.
    It first counts the frequency of each symbol in the sequence using the
    `Counter` class from the `collections` module. Then it creates a priority queue
    and inserts a node for each symbol into the queue. The nodes are then
    extracted two at a time, merged into a new node, and inserted back into
    the queue, until only one node remains. This node is the root of the Huffman tree.
    

Creates a Huffman tree from the given sequence.

    Parameters:
        sequence (str): The input sequence.

    Returns:
        Node: The root node of the Huffman tree.

    Algorithm:
        1. Count the frequency of each symbol in the sequence.
        2. Create a priority queue and insert nodes for each symbol with their frequencies.
        3. Repeat until only one node remains in the queue:
            a. Extract two nodes with the minimum frequencies.
            b. Merge the two nodes into a new node with the sum of their frequencies.
            c. Insert the merged node back into the priority queue.
        4. The remaining node in the queue is the root of the Huffman tree.

```python
def create_huffman_tree(sequence):
    # Liczenie częstości występowania symboli
    frequency = Counter(sequence)

    # Tworzenie kolejki priorytetowej
    pq = PriorityQueue()

    # Dodawanie wszystkich symboli do kolejki priorytetowej
    for symbol, freq in frequency.items():
        pq.insert(Node(freq, symbol))

    # Tworzenie drzewa Huffmana
    while len(pq.queue) > 1:
        left = pq.extract_min()
        right = pq.extract_min()
        merged = Node(left.freq + right.freq, left.symbol + right.symbol, left, right)
        pq.insert(merged)

    return pq.extract_min()
```

**generate_huffman_codes** his function generates the Huffman codes for each symbol in the Huffman tree.
    It performs a depth-first traversal of the tree, appending '0' to the code when
    moving to the left child and '1' when moving to the right child. When it reaches
    a leaf node (a node with no children), it adds the symbol and its code to the dictionary of codes.
    
Generates Huffman codes for each symbol in the Huffman tree.

    Parameters:
        node (Node): The current node in the Huffman tree.
        binary_code (str): The binary code obtained so far.
        code_dict (dict): Dictionary to store Huffman codes.

    Returns:
        dict: Dictionary containing Huffman codes for symbols.

    Algorithm:
        1. Traverse the Huffman tree recursively.
        2. At each node:
            a. If it's a leaf node, assign the binary code to the symbol.
            b. Otherwise, traverse left child with '0' appended to the binary code,
               and right child with '1' appended.
        3. Return the dictionary containing Huffman codes.


```python
def generate_huffman_codes(node, binary_code='', code_dict={}):
    if node is None:
        return

    # Jest liściem
    if node.left is None and node.right is None:
        code_dict[node.symbol] = binary_code
        return

    generate_huffman_codes(node.left, binary_code + '0', code_dict)
    generate_huffman_codes(node.right, binary_code + '1', code_dict)

    return code_dict
```

    Encodes a sequence using Huffman codes.

    Parameters:
        sequence (str): The input sequence to be encoded.
        code_dict (dict): Dictionary containing Huffman codes for symbols.

    Returns:
        str: Huffman-encoded sequence.

    Examples:
        >>> huffman_encoding("abracadabra", {'a': '0', 'b': '110', 'r': '10', 'c': '1110', 'd': '1111'})
        '01100101110110101100'

```python
def huffman_encoding(sequence, code_dict):
    return ''.join(code_dict[symbol] for symbol in sequence)
```

    Decodes a Huffman-encoded sequence using the Huffman tree.

    Parameters:
        encoded_sequence (str): The Huffman-encoded sequence.
        huffman_tree (Node): The root node of the Huffman tree.

    Returns:
        str: Decoded sequence.

    Examples:
        >>> huffman_decoding('01100101110110101100', root_node)
        'abracadabra'

```python
def huffman_decoding(encoded_sequence, huffman_tree):
    decoded_sequence = []
    current_node = huffman_tree
    for bit in encoded_sequence:
        if bit == '0':
            current_node = current_node.left
        else:
            current_node = current_node.right

        if current_node.left is None and current_node.right is None:
            decoded_sequence.append(current_node.symbol)
            current_node = huffman_tree

    return ''.join(decoded_sequence)
```

```python
import heapq
from collections import defaultdict, Counter
import matplotlib.pyplot as plt
import networkx as nx


# Funkcja do wizualizacji drzewa Huffmana
def plot_huffman_tree(node):
    def add_edges(graph, node, pos={}, x=0, y=0, layer=1):
        if node is None:
            return pos

        pos[node.symbol] = (x, y)
        if node.left:
            graph.add_edge(node.symbol, node.left.symbol)
            pos = add_edges(graph, node.left, pos, x - 1 / layer, y - 1, layer + 1)
        if node.right:
            graph.add_edge(node.symbol, node.right.symbol)
            pos = add_edges(graph, node.right, pos, x + 1 / layer, y - 1, layer + 1)

        return pos

    graph = nx.DiGraph()
    pos = add_edges(graph, node)
    labels = {node: node for node in graph.nodes()}

    plt.figure(figsize=(12, 8))
    nx.draw(graph, pos, labels=labels, with_labels=True, node_size=2000, node_color='skyblue', font_size=10,
            font_weight='bold', arrows=False)
    plt.show()


# Przykład użycia
if __name__ == "__main__":
    sequence = "Denys Example"

    # Tworzenie drzewa Huffmana
    huffman_tree = create_huffman_tree(sequence)

    # Generowanie kodów Huffmana
    huffman_codes = generate_huffman_codes(huffman_tree)
    print("Kody Huffmana dla wszystkich symboli:")
    for symbol, code in huffman_codes.items():
        print(f"{symbol}: {code}")

    # Kodowanie sekwencji
    encoded_sequence = huffman_encoding(sequence, huffman_codes)
    print("\nZakodowana sekwencja:")
    print(encoded_sequence)

    # Dekodowanie sekwencji
    decoded_sequence = huffman_decoding(encoded_sequence, huffman_tree)
    print("\nZdekodowana sekwencja:")
    print(decoded_sequence)

    # Wizualizacja drzewa Huffmana
    print("\nDrzewo Huffmana:")
    plot_huffman_tree(huffman_tree)
```
